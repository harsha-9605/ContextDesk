# AI Semantic Understanding Research
Testing if BERT can understand the semantic relationship between sentences.

```python
import sys
import os

# Add the backend folder to the Python path so we can import our modules
backend_path = os.path.abspath('../backend')
if backend_path not in sys.path:
    sys.path.append(backend_path)

from ai.engine import SemanticEngine

# Initialize the AI Engine
engine = SemanticEngine()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

sentence1 = "I need help with my college fees"
sentence2 = "Where can I find scholarship information?"

# Generate vector embeddings for both sentences
vec1 = engine.generate_embedding(sentence1)
vec2 = engine.generate_embedding(sentence2)

# Compute the mathematical angle (cosine similarity) between the two vectors
# Reshape is needed because cosine_similarity expects 2D arrays
similarity = cosine_similarity(np.array([vec1]), np.array([vec2]))[0][0]

print(f"Sentence 1: '{sentence1}'")
print(f"Sentence 2: '{sentence2}'")
print(f"\nSemantic Similarity Score (0 to 1): {similarity:.4f}")
```
